```python
# Find the  prime numbers from 1 to 100 using function
def prime_check(n:int)->bool:
    if n < 2:
        return False
    for i in range(2,n//2+1):
        if n%i==0:
            return False
    return True
for j in range(1,100):
    if prime_check(j):
        print(j,end = " ")
```
