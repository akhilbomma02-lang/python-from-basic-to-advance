```python
# check whether the number is divisible with 3 or not
num=int(input())
if num%3 == 0 and num%5 ==0:
    print(f"{num} is  divisible with 3 and 5")
else:
    print(f"{num} is not  divisible with 3 and 5")
```

```python
# check whether the number is perfect square or not
num=int(input())
val=num ** 0.5
print(val)
if val == int(val):
    print(f"{num} is perfect square")
else:
    print(f"{num} is not perfect square")
```
