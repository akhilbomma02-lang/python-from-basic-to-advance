```python
# NESTED IF ELSE
# check weather the number is even or odd include 0 as even using nested if
num=int(input())
if num>=0:
    if num %2==0:
         print(f"{num} is +ve Even")
    else:
        print(f"{num} is +ve odd")
else:
     if num %2==0:
         print(f"{num} is -ve Even")
     else:
          print(f"{num} is -ve Odd")
         




    
```
