```python
# Built in functions:The functions defination is provided by python,those functions called as Built in Functions
```

```python
#abs() : it returns the absulute value
print(abs(20-10))
print(abs(10-20))
```

```python
# round(val,pos):This function rounds the float value to particular positions
print(round(2.2589,2))
print(round(2.2539,2))
print(round(2.2539))
print(int(2.2539))
```

```python
#ord(char):It returns ASCII value of character
# chr(ascii_val):It returns the character for specified ASCII value
print(ord("A"))
print(chr(65))
```

```python
# sorted(iterable,reverse=False, key=None):
# sort the fruits based on the length of fruit name
fruits=["Mango","Apple","banana","kivi","Water Melon"]
res=sorted(fruits)
print(res)
print(sorted(fruits, reverse=True))
print(sorted(fruits, reverse=True, key=len))# it sorts the list based on length of fruits name
```

```python
# map(function,iterable)
# Find the square of all list of integers using map function
def square(x):
    return x**2
arr=[1,2,3,4,5]
res=(list(map(square,arr)))
print(res)

```

```python
# filter(function,iterable): It filters the iterable, if the function returns Not-None Those values will add to result
# Find even numbers from list of numbers
def check_even(x):
    if x % 2 ==0:
        return x
lst=[1,2,3,4,5,6]
res=list(filter(check_even, lst))
print(res)
```

```python
# Find even numbers from list of numbers
def check_even(x):
    if x % 2 ==0:
        return x
lst=[1,2,3,4,5,6]
res=list(map(check_even, lst))
print(res)
```

```python
# Find the 3's multiplyers for list of integers
def check_multiplyers(x):
    if x%3==0:
        return x
lst=[1,24,54,22,44,51,33,15]
res=list(filter(check_multiplyers,lst))
print(res)
```

```python
#enumerate(iterable):This function returns the list of tuple of index and item for iterable items
fruits=["Mango","Apple","banana","kivi","Water Melon"]
print(enumerate(fruits))
print(list(enumerate(fruits)))
for ind, val in enumerate(fruits):
    print(f":{ind} - {val} and length is {len(val)}")
```

```python
# Tuple unpacking
t = (1,2)
a,b =t
print(a,b)
```

```python
# zip(iterable1, iterable2, ...):It returns all iterables as single iterable of tuples
rool_no = [100,102,103,104,105]
names = ["Ravi","Raju","siva","Rani","Gopi"]
percentage = [90,67,96.8,67.65,99.99999]
print(list(zip(rool_no,names,percentage)))
              
```

```python
num=int(input())
res = ""
while num>0:
    bit = num%2
    num = num//2
    res = str(bit) + res
print(res)
    
```

```python
# number conversions
print(bin(10)) # string type of binary value
print(bin(10) [2:])
print(oct(10))
print(hex(11).upper()[2:])
      
```

```python
# any() : This Function returns True when any one iterable items return True
# all() : This Function returns True when all iterable items return True
arr=[1,2,3,4,0]
print(all(arr))
print(any(arr))
```

```python
percentage = [34,35,54,65,67]
for num in percentage:
    if num < 35:
        print("All students not passed")
        break
else:
    print("All students pass")
```

```python
percentage = [34,35,54,65,67]
lst=[]
for num in percentage:
    if num < 35:
        lst.append(False)
        
    else:
        print(True)
print(lst)
if all(lst):
    print("All students pass")
else:
    print("All students not passed")
```
