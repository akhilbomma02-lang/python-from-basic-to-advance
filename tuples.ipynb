# Dictionary keys are unique
# Dictionary does not support list as keys because list is mutable
# Dictionary values allows duplicates

```python
#
details={'name':"Akhil","age":20,"course":"Python"}
print(details)
```

```python
#Dictionary cant allow list as key
#If dictionary contains multiple items with key,In that case it updates first item key value with last defined item value
details={'name':"Akhil","age":20,"course":"Python","marks":[85,90,73],"age":50}
print(details)
```

```python
#access values by  using keys
details={'name':"Akhil","age":20,"course":"Python","marks":[85,90,73],"age":50}
print(details["course"])
print(details["age"])
#Updating the dictionary values
details["age"]+=1
details["course"]="java"#updating the value
details["location"]="kadapa"#adding new item in the dictinary
print(details)
```

### Dictionary Methods
- Get value
     -get(key,default_val):it returns the value if key exists in dictionary otherwise
  -default_value
  -update values and Add items to dictionary
   -update(new_dictionary):Dictionary Updates with New_dictionary items
- Remove items from dictionary
   -pop(key):it returns value and remove item if key is exists in dictionary otherwise "KeyError"
  -popitem():it returns tuple of key value pair and removes the last item from dictionary
  -Get Keys,Value and ithems from dictionary
  .keys():It returns list of dictionary keys
  .Values():It returns list of dictionary values
  .items():It returns list of tuple key-Value pairs 

```python
details={'name':"Akhil","age":20,"course":"Python","marks":[85,90,73],"age":50}
print(details.get("name"))
```

```python
details={'name':"Akhil","age":20,"course":"Python"}
d={"marks":[85,90,73],"age":50}
details.update(d)
print(details)
```

```python
details={'name':"Akhil","age":20,"course":"Python","marks":[85,90,73],"age":50}
print(details.pop("age"))
print(details)
print(details.popitem())
print(details)
```

```python
details={'name':"Akhil","age":20,"course":"Python","marks":[85,90,73],"age":50}
print(details.keys())
print(details.values())
print(details.items())
```
