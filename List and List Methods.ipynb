### list methods
- Element Adding Methods
    -  append(item):it adds item at end of the list
    - extand(itarable) : it adds all iterable elements at end of the list
    - insert(pos,item) : This Method adds the item at specified position 

```python
lst=[1,2,3,4]
lst.append(10)
print(lst)
lst.append("code")
print
```

-Element remove methods
 - remove(val) :it removes first occurence of value if value exists in list,otherwise it raises 'ValueError'
 - pop():it removes and returns the last of element of list
 - pop(pos):it removes and returns the specified position element from list
 - clear():it removes all elements from list
 - del:it python keyword,used to delete values.
 - 

```python
lst=[10,20,30,20,40]
lst.remove(20)
print(lst)
lst.pop()
```

# -index(val):it returns the index of first occurence of value if value exists in list otherwise ValueError
# count(val):it returns the total frequence of value

```python
lst=[10,20,30,20,40]
a=lst.index(20)
print(a)
lst.count(10)
```

# sort(reverse =false):it sorts the all list elements in ascending order,
#                   if reverse value is true is sorts in descending order.



```python
lst=[5,43,63,73,-2,4,-6]
lst.sort()
print(lst)
```

```python
lst.sort(reverse=True)
print(lst)
```
