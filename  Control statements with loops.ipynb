```python
# count how many consonents and how many vowels
a=input()
v=0
c=0
for char in a:
    if char.isalpha():
        if char in "aeiou":
            v=v+1
        else:
            c=c+1
print(c,v)
```

```python
# count the even num & odd num in a given list of elemnts with their even and odd numbers
lis=list(map(int,input().split()))
even=[]
odd=[]
for i in lis:
    if i%2==0:
        even.append(i)
    else:
        odd.append(i)
print(even)
print(odd)

```
