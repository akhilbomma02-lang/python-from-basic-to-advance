```python
# Enclosing scope
def outer(x):
    y=10 # Enclosing scope
    def inner():
        a=10 # local scope
        z=y+a
        print(f"This is local scope:{z}")
    inner()
    y=y+x
    print(f"This is enclosing scope:{y}")
n=5
outer(n)

    
```

```python
# Global scope
x=10 # Global scope
def outer():
    y=20 # Enclosed scope
    def inner():
        z=30 # Local scope
        res=z+y+x
        print(f"This is local scope:{res}")
    res=y+x
    inner()
    print(f"This is Enclosing scope:{res}")
outer()
```

```python
# Global scope
x=10 # Global scope
def outer():
    x=20
    x=x+1
    print("This is Local scope val:", x)
outer()
print("Thsi is Global Scope Val:",x)
```

```python
# Trying to update the global scope variable inside the local scope
x=10 # Global scope
def outer():
    global x
    x=x+1
    print("This is Local scope val:", x)
outer()
print("Thsi is Global Scope Val:",x)
```

```python
# Update the global scope variable inside the local scope by using 'global' keyword
x=10 # Global scope
def outer():
    y=20 # Enclosing scope
    def inner():
        nonlocal y
        z=30 # Local variable
        y=y+1
        print(f"This is Local Scope: {y}")
    y=y+x 
    inner()
    y=y+x
    print(f"This is Enclosing scope:{y}")
outer()

```
