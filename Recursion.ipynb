```python
# Find factorial of a number
def factorial(n):
    if n==1 or n==0:
        return 1
    return n*factorial(n-1)
factorial(6)
    
```

```python
# Find the fibonacii ofv a given number using recurission
def fibanacii(n):
    if n==1 or n==0:
        return 1
    return fibanacii(n-1) + fibanacii(n-2)
fibanacii(6)
    

```

```python
# Find the sum of n natural numbers using rescursion
def summ(n):
    if n==1:
        return 1
    return n+summ(n-1)
summ(10)
```

```python
# Find the product of a n natural using recursion
def product(n):
    if n==1:
        return 1
    return n*product(n-1)
product(5)
```

```python
# Find the binary number using recursion
def binary(n):
    if n == 0:
        return "0"
    if n==1:
        return "1"
    return binary(n//2) + str(n%2)
binary(2)
    
        
        
```

```python
# Find square of a number using lamba
square = lambda x: x**2
print(square(10))
```

```python
# sum of two numbers using lambda
add=lambda x,y:x+y
print(add(4,5))
```

```python
# Find square of all list of integers using lamba and map()
arr=[1,2,3,4,5]
res=list(map(lambda x:x**2,arr))
print(res)
```

```python
# Find even numbers in list using lambda and filter()
arr=[1,2,3,4,5]
res=list(filter(lambda x:x%2==0,arr))
print(res)
    
```

```python
# single line if-else syntax
# True_exp if condition else False_exp
# Find the weather the number is even or odd using lambda
n = int(input())
res=lambda x:f"{x} odd" if x%2 else f"{x} - even"
print(res(n))
```
