```python
# print the given number is prime or not
n=int(input())
s=[]
for i in range(1,n+1):
    if n%i==0:
        s.append(i)
print(s)
if len(s)==2:
    print(f"{n} is Prime")
else:
    print(f"{n} is Not Prime")
    
```

```python
# print the given number is prime or not using for else loop
num=int(input())
for i in range(2,num//2+1):
    if num%i==0:
        print(f"{num} is  not prime")
        break
else:
    print(f"{num} is prime")
        
    
    
```

```python
# reverse of a string and print whether it is palindrome or not
S=input()
rev=""
for char in S:
    rev=char+rev
print(rev)
if rev==S:
    print(f"{S} is Palindrome")
else:
    print(f"{S} is Not Palindrome")
```

```python
# Write a python program to print the following alphabet pattern using nested for loops and the chr() function
n=4
for i in range(n):
    for j in range(n):
        print(chr(65+i), end=" ")
    print()

```

```python
# Write a Python program to print consecutive ASCII characters in a 7×7 matrix using nested loops and the chr() function.
n=7
k=0
for i in range(n):
    for j in range(n):
        print(chr(65+k), end =" ")
        k+=1
    print()
```
