```python
#### String Methods
- Case Conversion Methods
   - upper():it converts entire string into lowercase
   - lower(): it converts entire string into uppercase
   - title() :it converts 1st letter of each word into upper case
   - swapcase():it converts entire string as upper case to lower and lower to upper
   - capitalize():it converts the 1st letter uppercase of word in entire string
```

```python
string="The Ocean is blue and sky is Blue"
lower=string.lower()
print(lower)
print(string.upper())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

```python
# white space remove methods
# strip():It removes the white space to both sides of string
# lstrip:it removes the left side of the white space of string
# rstrip:it removes the right side of thye white space of string
string="        codegnan     "
print(string)
print(string.strip())
print(string.lstrip())
print(string.rstrip())
```
