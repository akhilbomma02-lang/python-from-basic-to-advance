### SET METHODS
- Adding Elements to set
   - add(ele):it adds element to set if element not exists in set,otherwise nothing
   - update(iterable):it adds all iterable elements to set
- Remove elemnts from set
   - remove(ele):it removes the element from set if element exists in set otherwise
-KeyError
    -discard(ele):it removes the elements from set if element exists in set otherwise nothing
- set operations
  - pop():it removes random elements from the set
  - union(another_sets):it returns the all common elements from all sets
  - intersection(another_sets):it returns the common elements from the all sets
  - difference(another_set): it returns the set elements those are not lies in another sets
  - symmetric_ difference(another_sets): it returns th Non-common elements from all


```python
s={10,40,30,60,45}# adding elements
s.add(80)
print(s)
```

```python
s={10,40,30,60,45}# removing elements from set
s.remove(30)
print(s)
s.remove(80)
print(s)
```

```python
s={10,40,30,60,45}
s.discard(46)
print(s)
```

```python
s1={1,2,5,7} # INTERSECTION
s2={3,4,5,6,7}
s3={6,8}
print(s1 & s2)
```

```python
s1={1,2,5,7}# difference OPERATOR
s2={3,4,5,6,7}
s3=(6,8)
print(s1 - s2 )
```
