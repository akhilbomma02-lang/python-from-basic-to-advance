```python
# 1. Using concatination
name="Ravi"
age=20
# My name Ravi and i am a 20 year old
print("My name " + name + " and iam a " + str(age) + " Year old")
```

```python
# 2. Access specifiers
# integer -%d
# float - %f
# string -%s
# 1. Access Specifiers 
name="Ravi"
age=20
marks=[10,20,30,40]
print("My name %s and i am a %d year old " % (name,age))
```

```python
# 3. Dot format
name="Ravi"
age=20
print("My name {} and i am a {} year old".format(name,age))
```

```python
# 4.f - format
name="Ravi"
age=20
percentage=98.5878
print(f"My name {name} and i am a{age} years old and my percemtage is {percentage:.2f}") 
```
