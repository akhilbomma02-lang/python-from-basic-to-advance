```python
# Find weather the number is number is prime number or not using function
def prime_check(n:int)->bool:
    for i in range(2,n//2+1):
        if n%i==0:
            return False
    return True
num=int(input())
print(prime_check(num))
            
            
    


```

```python
# Find the  prime numbers from 1 to 100 using function
def prime_check(n:int)->bool:
    if n < 2:
        return False
    for i in range(2,n//2+1):
        if n%i==0:
            return False
    return True
for j in range(1,100):
    if prime_check(j):
        print(j,end = " ")
```

```python
# Types of Arguments
# 1. positional Arguments: The arguments are assigns to parameters as per the positional order

def sub(x:int, y:int) ->int:
    c=x - y
    return c
# functional calling block
a,b=map(int,input().split())
res=sub(a,b)
print(res)
print(sub(b,a))
```

```python
# Types of Arguments
# 2. Keyword Arguments:

def sub(x:int, y:int) ->int:
    c=x - y
    return c
# functional calling block
a,b=10,20
print(sub(x=a,y=b))
print(sub(y=b,x=a))

      
```

```python
# 2 numbers sum
def add(x:int, y:int) ->int:
    c=x + y
    return c
# functional calling block
a,b=10,20
print(add(x=a,y=b))
print(add(y=b,x=a))

    
```

```python
# 3 numbers sum
def add(x:int, y:int,z:int) ->int:
    c=x-x+y+z
    return c
# functional calling block
a,b,c=10,20,30
print(add(x=a,y=b,z=c))
print(add(y=b,x=a,z=c))
```

```python
# Default Arguments:The parameters defined with some default values, and its assigns 
# the argument values if you pass the values for default paraments otherwise,it will assign default values 
def add(a,b,c=0,d=0):
    result=a+b+c+d
    return result
print(add(10,20))
print(add(10,20,5))
print(add(10,20,2,1))
```

```python
def mul(a,b,c=1,d=1):
    result=a*b*c*d
    return result
print(mul(10,20))
print(mul(10,20,5))
print(mul(10,20,2,1))
```

```python
# 4. Variable length of positional Arguments
def add(a,b=0,*args):
    print(type(args))
    print(args)
    s=0
    for num in args:
        s+=num
    return s+a+b
print(add(1,2,3,4,5,6,7,8,9))
```

```python
# variable Length of Keyword Arguments
def details(name,age,*marks,**kwargs):
    print(type(kwargs))
    print(kwargs)
    print(kwargs.keys())
print(details("Ravi",20,85,95,75,gender="Male",location="India",percentage=95.8))
    
```

```python
# pass by value 
def fun(x):
    print(f"Before update Address of x :{id(x)} and value: {x}")
    x=x+1
    print(f"After update the x value, x:{id(x)} and value: {x}")
a=5
print(f"Before function execution address of a:{id(a)} and value:{a}")
print(fun(a))
print(f"After function execution address of a:{id(a)} and value:{a}")
```

```python
# pass by Refference 
def fun(x):
    print(f"Before update Address of x :{id(x)} and value: {x}")
    x.append(100)
    print(f"After update the x value, x:{id(x)} and value: {x}")
a=[1,2,3]
print(f"Before function execution address of a:{id(a)} and value:{a}")
print(fun(a))
print(f"After function execution address of a:{id(a)} and value:{a}")
```
