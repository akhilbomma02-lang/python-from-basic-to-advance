```python
# remove the target element from the given list using for loop
lst=list(map(int,input().split()))
a=[]
Target=int(input())
for ele in lst:
    if ele!=Target:
        a.append(ele)
print(a)
```

```python
# remove the target element from the given list using while loop
lst=list(map(int,input().split()))
a=[]
target=int(input())
i=0
n=len(lst)
while i<n:
    if lst[i]!=target:
        a.append(lst[i])   
    i=i+1
print(a)
    
```

```python
# remove the target element from the given list withou using extra space
lst=[1,2,3,4,5,6,3,9,2,3]
target=3
n=len(lst)
i=0
while i<n:
    if lst[i]==target:
        lst.pop(i)
        n=n-1
    i=i+1    
print(lst)
    
        
    
```

```python
# find the target index number using break statement:
a=[1,4,5,6,9]
target=5
b=len(a)
for i in range(b):
    if a[i]==target:
        print(i)
        break
else:
    print("-1")
        
```
