```python
# Print a 5*5 square star(*) pattern
n=5
for i in range(n):
    for j in range(n):
        print("*", end=" ")
    print()
```

```python
# print number from 1 to 25 in 5*5 square pattern
n=5
for i in range(n):
    for j in range(n):
        print(n*i+j+1, end="\t ")
    print()
```

```python
#Print number 1 to 5 repeatedly in 5 rows
n=5

for i in range(n):
    for j in range(n):
        print(j+1, end="\t ")
    print()
```

```python
# write a python program to print a 4*4 checkerboard pattern using 0s and 1s.
n=4
for i in range(n):
    for j in range(n):
        print((i+j)%2, end="\t ")
    print()
```

```python
# write aa python program to print a right aligned triangle using '*' symbols
n=5
for i in range(n):
    for j in range(n):
        if j>= n-i-1:
            print("*",end=" ")
        else:
            print(" ", end=" " )
    print()
```

```python
# write a python program to print left diagonal (*) pattern
n=5
for i in range(n):
    for j in range(n):
        if i==j:
            print("*",end=" ")
        else:   
            print(" ",end=" " )
    print()
```

```python
# wwrite the python [programme to print the letter 'X' using * symbol 
n=5
for i in range(n):
    for j in range(n):
        if i==j or j==n-i-1:
            print("*",end=" " )
        else:
            print(" ",end=" " )
            
    print()
```

```python
# write a programme to print the inverted upper triangle using "*" pattern
n=5
for i in range(n):
    for j in range(n):
        if i<=j:
            print("*",end=" " )
        else:
            print(" ",end=" " )
            
    print()
```

```python
# write a python program to print hollow sqaure using '*' symbols
n=5
for i in range(n):
    for j in range(n):
        if i==0 or j==0 or i==n-1 or j==n-1:
            print("*",end=" " )
        else:
            print(" ",end=" ")
    print()
```
